```python
def rmse(données):
    rmse_t0 = np.sqrt(((données["out_ghi_t0"] - données["target_ghi_t0"])**2).sum()/len(données))
    rmse_t1 = np.sqrt(((données["out_ghi_t1"] - données["target_ghi_t1"])**2).sum()/len(données))
    rmse_t3 = np.sqrt(((données["out_ghi_t3"] - données["target_ghi_t3"])**2).sum()/len(données))
    rmse_t6 = np.sqrt(((données["out_ghi_t6"] - données["target_ghi_t6"])**2).sum()/len(données))
    print(rmse_t0, rmse_t1, rmse_t3, rmse_t6)
    print("mean=", np.mean([rmse_t0, rmse_t1, rmse_t3, rmse_t6]))
```

```python
def target_to_category(row):
    if row["night"]==1:
        return "night"
    if row["cloudy"]==1:
        return "cloudy"
    if row["slightly_cloudy"]==1:
        return "slightly_cloudy"
    if row["clear"]==1:
        return "clear"
    if row["variable"]==1:
        return "variable"
    
def output_to_category(row):
    if row["out_night"]==1:
        return "night"
    if row["out_cloudy"]==1:
        return "cloudy"
    if row["out_slightly_cloudy"]==1:
        return "slightly_cloudy"
    if row["out_clear"]==1:
        return "clear"
    if row["out_variable"]==1:
        return "variable"
```

```python
données[target_cloudiness + one_hot].head(10)
df=données

df["target_category"] = df.apply(target_to_category,axis=1)
df["output_category"] = df.apply(output_to_category,axis=1)
```
